```python
# Papagaio

# features [ 1 Sim, 0 No]
# tem bico?
# risco de extino?
# faz cocoric?

papagaio1 = [1,1,0]
papagaio2 = [1,1,0]
papagaio3 = [1,1,1]
```

```python
# Galinha

# features [1 sim, 0 no]
# tem bico?
# risco de extino?
# faz cocoric?

galinha1 = [1,0,1]
galinha2 = [1,0,1]
galinha3 = [0,0,0]
```

```python
dados = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
classes = [1, 1, 1, 0, 0, 0]
dados
```

```python
! pip install scikit-learn
```

```python
from sklearn.svm import LinearSVC

modelo = LinearSVC()
```

```python
modelo.fit(dados, classes)
```

```python
# features [1 Sim, 0 No]
# tem bico?
# risco de extino?
# faz cocoric?

# tem bico, est em risco de extino e no faz cocoroc

animal_misterioso = [1, 1, 0]
previso = modelo.predict([animal_misterioso])
previso
```

```python
# features [1 Sim, 0 No]
# tem bico?
# risco de extino?
# faz cocoric?

# tem bico, est em risco de extino e no faz cocoric

animal_misterioso2 = [1, 0, 1]
previsao2 = modelo.predict([animal_misterioso2])

previsao2
```

## Calculando a Acurcia do Modelo

## Dados e Rtulos

```python
# dados de treino x
treino_x = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
# roltulo (classe) de treino_y
treino_y = [1, 1, 1, 0, 0, 0,]
```

```python
# animal misteriosos - para teste
misterio1 = [1, 1, 1]
misterio2 = [1, 1, 0]
misterio3 = [0, 1, 1]
```

```python
# previses - dados de testes
teste_x = [misterio1, misterio2, misterio3]
# prevendo os testes
previsoes = modelo.predict(teste_x)
```

```python
# rtulos (classes) de teste
teste_y = [0, 1, 1]
```

```python
# importando mdulos que j traz a funo de acurcia encapsulada
from sklearn.metrics import accuracy_score

# definindo uma variavel para o retorno da funo acurcia * 100 para porcentagem
taxa_de_acerto = accuracy_score(teste_y, previsoes) * 100
```

```python
# mostrando a acurcia
print(f"Acurcia: {taxa_de_acerto:.2f}%")
```
